#### C√°lculo da M√©dia
<hr style="border: 1px solid rgba(242, 248, 241, 1); width: 100%; margin: 5px 0;">


A f√≥rmula da **m√©dia** √© usada para calcular o valor m√©dio de um conjunto de dados. A f√≥rmula √© dada por:

$$
\text{M√©dia} = \frac{\sum_{i=1}^{n} x_i}{n}
$$

Onde:
- $( x_i )$ representa cada valor do conjunto de dados.
- $( n )$ √© o n√∫mero total de elementos.

### Exemplo:

Se tivermos os seguintes n√∫meros: **2, 4, 6, 8, 10**, a m√©dia ser√°:

$$
\text{M√©dia} = \frac{2 + 4 + 6 + 8 + 10}{5} = \frac{30}{5} = 6
$$

```python
# M√©dia

a = [2, 4, 6, 8, 10]

print(f'M√©dia: {sum(a) / len(a)}')
```

```python
import statistics as st

print(f'M√©dia: {st.mean(a)}')
```

```python
# n√∫mero √≠mpar de elementos
b = [1500, 1800, 2000, 2100, 2200, 2400]

print(f'Mediana: {sorted(b)[len(b) // 2] if len(b) % 2 != 0 else (sorted(b)[len(b) // 2 - 1] + sorted(b)[len(b) // 2]) / 2}')
```

```python
import numpy as np
import pandas as pd
from statistics import mode

notas = [8, 7, 6, 7, 7, 8, 10, 9, 6, 7, 8, 2, 10, 9, 7, 6, 8, 9, 5, 10]

media = np.mean(notas)
mediana = np.median(notas)
moda = mode(notas)

print("M√©dia   :", media)
print("Mediana :", mediana)
print("Moda    :", moda)
```

```python
amplitude = max(notas) - min(notas)
variancia = np.var(notas, ddof=1)   # ddof=1 ‚Üí amostral
desvio_padrao = np.std(notas, ddof=1)
cv = desvio_padrao / media

print("Amplitude        :", amplitude)
print("Vari√¢ncia        :", variancia)
print("Desvio Padr√£o    :", desvio_padrao)
print("Coef. Varia√ß√£o   :", cv)
```

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)

# Par√¢metros
n_customers = 1500
start = pd.Timestamp("2023-01-01")
end   = pd.Timestamp("2025-07-31")
dates = pd.date_range(start, end, freq="D")

# Tabela de clientes
plans = ["Basic", "Pro", "Enterprise"]
regions = ["Sudeste", "Sul", "Nordeste", "Centro-Oeste", "Norte"]

customers = pd.DataFrame({
    "customer_id": np.arange(1, n_customers+1),
    "signup_date": rng.choice(pd.date_range("2023-01-01","2025-06-01", freq="D"), size=n_customers),
    "plan": rng.choice(plans, p=[0.55, 0.35, 0.10], size=n_customers),
    "region": rng.choice(regions, size=n_customers, p=[0.45,0.20,0.18,0.10,0.07])
})

# Probabilidade de churn depende do plano (mais alta no Basic)
base_churn = customers["plan"].map({"Basic":0.22,"Pro":0.12,"Enterprise":0.06}).values
churned = rng.random(n_customers) < base_churn
customers["churned"] = churned
# churn_date ap√≥s signup para quem churnou
churn_dates = customers["signup_date"] + pd.to_timedelta(rng.integers(30, 540, size=n_customers), unit="D")
customers["churn_date"] = np.where(churned, churn_dates, pd.NaT)
customers["churn_date"] = pd.to_datetime(customers["churn_date"])

# Cat√°logo simples
products = pd.DataFrame({
    "product": ["Widget A","Widget B","Gadget X","Gadget Y","Service Plus","Service Care","Add-on 1","Add-on 2"],
    "category": ["Hardware","Hardware","Hardware","Hardware","Servi√ßo","Servi√ßo","Add-on","Add-on"],
    "base_price": [120, 200, 340, 280, 150, 80, 40, 60]
})

# Gera√ß√£o de pedidos (vendas)
orders_list = []
for d in dates:
    # sazonalidade simples: mais vendas em fim de m√™s e em nov-dez
    month = d.month
    dow = d.dayofweek
    mult = 1.0 + 0.25*(month in [11,12]) + 0.15*(d.day>25) + 0.10*(dow in [4,5])  # 6=s√°bado, 0=segunda
    n_orders = rng.poisson(80*mult)

    cust_today = rng.choice(customers["customer_id"].values, size=n_orders, replace=True)
    prod_today = rng.choice(products["product"].values, size=n_orders, replace=True)

    df_day = pd.DataFrame({"order_date": d, "customer_id": cust_today, "product": prod_today})
    df_day = df_day.merge(products, on="product", how="left")
    # quantidade e pre√ßo com leve varia√ß√£o
    df_day["qty"] = rng.integers(1, 5, size=n_orders)
    df_day["unit_price"] = (df_day["base_price"] * rng.normal(1.0, 0.08, size=n_orders)).round(2)
    df_day["revenue"] = (df_day["qty"] * df_day["unit_price"]).round(2)
    orders_list.append(df_day)

orders = pd.concat(orders_list, ignore_index=True)

# Junta metadados do cliente (plan, region) no pedido
orders = orders.merge(customers[["customer_id","plan","region","signup_date","churned","churn_date"]],
                      on="customer_id", how="left")

# Filtra pedidos ap√≥s signup e antes do churn (se houver)
orders = orders[(orders["order_date"] >= orders["signup_date"]) &
                (orders["churn_date"].isna() | (orders["order_date"] <= orders["churn_date"]))].reset_index(drop=True)

orders.head(), customers.head()
```

```python
import matplotlib.pyplot as plt

rev_daily = orders.groupby("order_date")["revenue"].sum().rename("revenue")
rev_7dma = rev_daily.rolling(7).mean()

plt.figure(figsize=(10,4))
plt.plot(rev_daily.index, rev_daily.values, label="Receita di√°ria")
plt.plot(rev_7dma.index, rev_7dma.values, label="M√©dia m√≥vel 7d")
plt.title("Receita ao longo do tempo")
plt.xlabel("Data"); plt.ylabel("Receita")
plt.legend(); plt.tight_layout(); plt.show()
```
